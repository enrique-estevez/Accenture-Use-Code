This Notebook intends to serve as a guide for understanding the ADF test from scratch. It combines the theory and mathematical background required to understand the test as well as code snippets exemplifying the use of each feature.

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
```

## 1. Dickey-Fuller test
The augmented Dickey-Fuller (ADF) test is an extension of the Dickey-Fuller test. Therefore, we need to better understand the latter before moving on to the ADF test. This test was first proposed by **David Dickey** and **Wayne Fuller** in **1979** as a way to check wheter a time series is stationary.

This test falls within the category of unit root tests. To better understand this concept we need to first define the concept of autoregressive (AR) model. An AR is a representation of a process where the output variable depends **linearly on its own previous values** and on a stochastic term. The number of previous values or **lags** considered when estimating the output variable is denominated the **order** of the model. For instance an **autoregressive model of order 1** or **AR(1)** can be defined as follows:

$y_t = \rho y_{t-1}+u_t$ ;  where $u_t$ represents a white noise process.

In this example, we speak of time series having a unit root when $\rho=1$, since this makes 1 a root of the characteristic equation of the process. This is equivalent to stating that the process is **integrated of order 1** or **I(1)**. When a unit root is present in a process, this process is **non-stationary** or a **random walk** as opposed to a **causal process** or **stationary process**.

This is precisely the Dickey-Fuller test's goal, to check wheter a unit root is present in the time series under analysis. To do so, the following hypothesis testing is performed:

Given: $y_t = \rho y_{t-1}+u_t$

$\begin{cases}H_0: \rho=1 \\H_1: \rho\neq1 \end{cases}$

More often, the AR(1) process is rewritten to the following equivalent expression:

$y_t=\rho y_{t-1}+u_t \Leftrightarrow y_t - y_{t-1}=\rho y_{t-1}-y_{t-1}+u_t \Leftrightarrow \Delta y_t=(\rho-1)y_{t-1}+u_t \Rightarrow $ 

$\Rightarrow \Delta y_t=\delta y_{t-1}+u_t$ ;  where $\delta=(\rho-1)$ and $u_t$ represents a white noise process.

Under this formulation the hypothesis testing is presented as: $\begin{cases}H_0: \delta=0 \\H_1: \delta\neq0 \end{cases}$

### 1.1 Specification of the DF statistic
In order to test the above hypothesis, we need to develop a test statistic that hopefully is asymptotically normal. We can do that by using the least square estimator of $\rho$. To compute the least square estimator of $\hat{\rho}$ we first derive the expression of the estimation error of $y_t$ with respect to $\hat{\rho}$ and then we set this expression to zero and solve for $\hat{\rho}$.

The expression for the sum of the squared errors of the model is: $\displaystyle SSE=\sum^n{(y_t-\hat{\rho}y_{t-1})^2}$

Then we take the partial derivative with respect to $\hat{\rho}$ and set it to zero:

$\displaystyle\frac{\partial SSE}{\partial \hat{\rho}}=-2\sum^n{\Big(y_t-\hat{\rho}y_{t-1}\Big)y_{t-1}}=-2\sum^n{\Big(y_ty_{t-1}-\hat{\rho}y_{t-1}^2\Big)}=0$

Solving for $\hat{\rho}$ we get: $\displaystyle\hat{\rho}=\frac{\sum^n{y_ty_{t-1}}}{\sum^n y_{t-1}^2}$

Now we have to do some algebra to transform this expression into a distribution of the test statistic. We begin by multiplying numerator and denominator by $\frac{1}{n}$ and reordering terms considering $y_t=y_{t-1}+u_t$. Then we divide both numerator and denominator by $\sigma_{e}^2$.

$\hat{\rho}=\displaystyle\frac{\sum{y_ty_{t-1}}}{\sum y_{t-1}^2}=\displaystyle\frac{\frac{1}{n}\sum{y_ty_{t-1}}}{\frac{1}{n}\sum y_{t-1}^2}=\displaystyle\frac{\frac{1}{n}\sum{(y_{t-1}+u_t)y_{t-1}}}{\frac{1}{n}\sum y_{t-1}^2}=1+\displaystyle\frac{\frac{1}{n}\sum{u_ty_{t-1}}}{\frac{1}{n}\sum y_{t-1}^2}=1+\displaystyle\frac{\frac{1}{n\sigma_{e}^2}\sum{u_ty_{t-1}}}{\frac{1}{n\sigma_{e}^2}\sum y_{t-1}^2}$ 

We will analyze the above expression by parts.<br/><br/>


**Numerator**

Under $H_0$ we have $y_t=y_{t-1}+u_t$. Therefore, squaring both sides of the expression we get $y_t^2=y_{t-1}^2+2y_{t-1}u_t+u_t^2$. 

Regrouping terms, $y_{t-1}u_t=\frac{1}{2}(y_t^2-y_{t-1}^2-u_t^2)$. 

Hence, $\displaystyle\frac{1}{n\sigma_{e}^2}\sum^n{u_ty_{t-1}}=\frac{1}{2}\Bigg(\frac{\sum^n(y_t^2-y_{t-1}^2)}{n\sigma_{e}^2}-\frac{\sum^n u_n^2}{n\sigma_{e}^2}\Bigg)=\frac{1}{2}\Bigg(\frac{y_n^2}{n\sigma_{e}^2}-\frac{\sum^n u_n^2}{n\sigma_{e}^2}\Bigg)$

We know $y_n=y_{n-1}+u_t=y_{n-2}+u_{t-1}+u_t=\sum^n u_n$, therefore $Y_n\sim N(0,n\sigma^2_e)$ where $N(\mu, \sigma^2)$ represents a gaussian distribution of mean $\mu$ and variance $\sigma^2$. Hence, $\displaystyle\frac{Y^2_n}{n\sigma^2_e}\sim \chi^2_1$ where $\chi^2_1$ represents a chi-squared distribution with one degree of freedom. It is important here to remember that the square of a brownian motion with t steps converges in distribution to t times a chi-squared distribution with one degree of freedom; that is, $B^2(t)\overset{d}{\longrightarrow}t\chi^2_1$.

In the same way $\displaystyle\sum^nu_n\overset{p}{\longrightarrow}\sigma^2_e$ since the average of a white noise process converges in probability to the variance of the process. Then $\displaystyle\frac{1}{n\sigma^2_e}\sum^nu_n\overset{p}{\longrightarrow}1$

Putting all together, the numerator can be expressed as $\displaystyle\frac{1}{2}(\chi^2_1-1)$.<br/><br/>


**Denominator**\
Considering the central limit theorem we know that $\displaystyle\frac{1}{n}\sum^n_{t-1}\bigg(\frac{y_{t-1}}{\sigma_e\sqrt{n}}\bigg)^2\overset{d}{\longrightarrow}\int_0^1B^2(t)dt$ where $B(t)$ represents a brownian motion process satisfying:\
(i)$\ \ \ \ B(0) = 0$\
(ii)$\ \ \{B(t_2)-B(t_1),...,B(t_n)-B(t_{n-1})\}$ are independent\
(iii) $\ \ B(t+\Delta(t))-B(t)\sim N(0, \Delta(t))$ with $\Delta(t)>0$.<br/><br/>

**Distribution for the statistic**
Putting together the numerator and the denominator we can conclude that the test statistic converges to the following distribution: $\displaystyle(\rho-1){\longrightarrow}\frac{\chi^2(1)-1}{2\int_0^1B^2(t)dt}$. This distribution do not have a closed form, so it has to be computed numerically. We will do so by using a Monte-Carlo simulation.

```python
def dickey_fuller_simulation(n: int) -> float:
    """
    This function generates a Montecarlo simulation following the Dickey-Fuller 
    distribution based on a number of time steps.
    
    The simulated distribution of the Dickey-Fuller statistic is: (χ2[1]-1) / (2 * √∫χ2[t]dt), 
    where χ2[n] represents the chi-squared distribution with n degrees of freedom.
    
    :param n: number of time steps for the simulation.
    :return: simulated value of the Dickey-Fuller statistic.
    """
    # Generate the series of white noise
    u_t = np.random.normal(loc=0, scale=1, size=n)
    # Acumulate the noise to simulate y_t
    y_t = np.cumsum(u_t)
    # Calculate the chi-squared distribution
    chi_sq = (1 / n) * y_t**2
    # Calculate the DF statistic value
    df_statistic = (chi_sq[-1] - 1)  /  (2*(np.mean(chi_sq))**0.5)
    return df_statistic
    
    
n = 5_000
simulations = 100_000
df_statistics = []

for i in range(simulations):
    df_statistics.append(dickey_fuller_simulation(n))
df_statistics.sort()

results = pd.DataFrame(0, index=['Expected', 'Actual'], columns=['1.0%', '2.5%', '5.0%', '10%'])
results.loc[:, '1.0%'] = -2.567, round(np.quantile(df_statistics, 0.01), 3)
results.loc[:, '2.5%'] = -2.223, round(np.quantile(df_statistics, 0.025), 3)
results.loc[:, '5.0%'] = -1.941, round(np.quantile(df_statistics, 0.05), 3)
results.loc[:, '10%'] = -1.616, round(np.quantile(df_statistics, 0.1), 3)
display(results)
```

So far we have derived the distribution for the Dickey-Fuller statistic under the base model $y_t = \rho y_{t-1}+u_t$ where there are no constant or trend. However, we can similarly derive the models for the following cases:

· Constant: $y_t = \alpha_0+\rho y_{t-1}+u_t$\
· Constant and trend: $y_t = \alpha_0+\alpha_1t+\rho y_{t-1}+u_t$

Every version of the test has its own Dickey-Fuller distribution and critical values.

It is important to use prior knowledge or use the appropiate tests to decide what model to use, since it can heavily impact the quality of the fit. Not considering constant and trend when it should be considered may lead to bias in the estimated value of $\rho$. Incorrectly considering constant or trend whent these should not be included may lead to a reduction in the power of the test (the probability of the test rejecting $H_0$ when $H_1$ is true).

The Dickey-Fuller test is known to be flawed by what is called the **"near observation equivalence"** problem. This problem consists on the model not being able to distinguish between $\rho=0$ and $\rho\approx0$. This is also expressed as the Dickey-Fuller test having **low statistical power**.

### 1.2 Dickey-Fuller test in Python
In order to perform the Dickey-Fuller test on a time series, we need the **adfuller** function from the module **statsmodels**. In fact this function performs the Augmented Dickey-Fuller test, but we can force it to perform the simple Dickey-Fuller test by limiting the number of lags considered to zero via the argument **'maxlag'**. 

We can determine the specification of the model through the argument **'regression'** where **'nc'** stands for the baseline model without constant or trend, **'c'** stands for the model with constant, **'ct'** stands for the model with constant and trend and **'ctt'** stands for the model with constant and linear and quadratic trend (not considered in this analysis). 

```python
# Generate a random walk (y_t)
u_t = np.random.normal(loc=0, scale=1, size=5000)
y_t = np.cumsum(u_t)

# Perform the Dickey-Fuller test
df_test = adfuller(y_t, maxlag=0, regression='nc', autolag=None)
print('Dickey-Fuller test statistic:    ', round(df_test[0],3))
print('MacKinnon’s approximate p-value:  ', round(df_test[1], 2))

# Plot the time series
plt.figure(figsize=(8, 5))
plt.title('Random walk example')
plt.plot(y_t)
plt.show();
```

## 2. Augmented Dickey-Fuller test
The Augmented Dickey-Fuller test (ADF test) is merely an extension of the baseline Dickey-Fuller test to consider **higher order auto-regressive processes**. The full model with constant and linear trend under the ADF test is the following:

$\Delta y_t = \alpha_0+\alpha_1t+\delta y_{t-1}+\phi_{1}\Delta y_{t-1}+...+\phi_{p}\Delta y_{t-p}+u_t$

The order $p$ of the process determines the higher auto-regressive term to be included in the model. This a parameter that must be adjusted before selecting the model. Usually, several models with different $p$ values are fit to adjust the optimum value.

It is worth noting that even considering the higher order auto-regressive terms, **the hypothesis testing is the same** that under the DF test, and the same statistic distributions hold:

$\begin{cases}H_0: \delta=0 \\H_1: \delta\neq0 \end{cases}$

Also, the same cautions regarding the model selection and the low statistical power of the test apply.

### 2.1 Augmented Dickey-Fuller test in Python
In order to perform the Dickey-Fuller test on a time series, we need the same **adfuller** function from the module **statsmodels** we used before.

Same as with the Dickey-Fuller case, we can determine the specification of the model through the argument **'regression'** where **'nc'** stands for the baseline model without constant or trend, **'c'** stands for the model with constant, **'ct'** stands for the model with constant and trend and **'ctt'** stands for the model with constant and linear and quadratic trend (not considered in this analysis). 

The order $p$ of the process can be fixed based on prior knowledge or can be optimized by the **adfuller** function. The behavior of the function can be adjusted via the **autolag** parameter. If we want to set a fixed maximum order value, we must set the **maxlag** value to this $p$ and the **autolag** parameter to None for the function to take the order as given by **maxlag**. 

If we want the function to find the optimum $p$ for us, we can use different optimization methods by changing the **autolag** configuration. The default method is the **'AIC'** (Akaike Information Criterion). Similarly, we can set **autolag** to **'BIC'** (Bayesian Information Criterion) to use this information metric. The last methodology for determining the optimum value of $p$ is using the **'t-stat'** method. Under this configuration the library starts fitting the ADF models from the **maxlag** value of $p$, droping one lag at each step, and stops when the t-statistic of the last lag reaches statistical significance at 5% level.

It is worth noting that using an automated method to find the optimum number of lags heavily impacts the execution time of the program, since multiple models must be fitted and evaluated.

```python
# Generate a random walk (y_t)
u_t = np.random.normal(loc=0, scale=1, size=5000)
y_t = np.cumsum(u_t)

# Add some auto-regressive terms up to order p=2
y_t[1:] = y_t[1:] + u_t[:-1]
y_t[2:] = y_t[2:] + u_t[:-2]

# Perform the Augmented Dickey-Fuller test
adf_test = adfuller(y_t, maxlag=50, regression='nc', autolag='BIC')
print('Optimum auto-regressive order:    ', adf_test[2])
print('Dickey-Fuller test statistic:    ', round(adf_test[0],3))
print('MacKinnon’s approximate p-value:  ', round(adf_test[1], 2))

# Plot the time series
plt.figure(figsize=(8, 5))
plt.title('Random walk example')
plt.plot(y_t)
plt.show();
```
